```python
def add(x, y):
    return x + y
```

```python
def subtract(x, y):
    return x - y
```

```python
def multiply(x, y):
    return x * y
```

```python
def divide(x, y):
    return x / y
```

```python
# Vamos testar cada função
print(f"{1} + {2} = {add(1, 2)}")
print(f"{10} - {4} = {subtract(10, 4)}")
print(f"{5} * {7} = {multiply(5, 7)}")
print(f"{12} / {6} = {divide(12, 6)}")
```
